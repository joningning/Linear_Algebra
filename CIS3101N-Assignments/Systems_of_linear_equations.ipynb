```python
import numpy as np

def RowSwap(A,k,l):
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A
    
    B = np.copy(A).astype('float64')
        
    for j in range(n):
        temp = B[k][j]
        B[k][j] = B[l][j]
        B[l][j] = temp
        
    return B

def RowScale(A,k,scale):
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A
    
    B = np.copy(A).astype('float64')

    for j in range(n):
        B[k][j] *= scale
        
    return B

def AddRow(A,k,l,scale):
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A
    
    B = np.copy(A).astype('float64')
        
    for j in range(n):
        B[l][j] += B[k][j]*scale
        
    return B
```

```python
def backwardSubstitution(U, b):
    n = U.shape[0]
    retnum = np.zeros(n) 

    for i in range(n - 1, -1, -1):
        retnum[i] = b[i]
        for j in range(i + 1, n):
            retnum[i] -= U[i][j] * retnum[j]
        if U[i][i] != 0:
            retnum[i] /= U[i][i]
        else:
            retnum[i] = 0  

    return retnum
```
